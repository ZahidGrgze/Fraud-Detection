```python
def plot_histogram_with_skew(df):
    num_variables = df.shape[1]

    # 7x4'lük bir ızgara oluştur
    fig, axes = plt.subplots(7, 4, figsize=(15, 20))  # Boyutları ayarlayabilirsiniz
    axes = axes.ravel()  # 2D axes listesini 1D yapar

    for ax, (column, values) in zip(axes, df.iteritems()):
        sns.histplot(values, ax=ax, kde=True)
        skewness = skew(values)

        # Çarpıklık değerini grafiğin üstüne yazdır
        ax.annotate(f'Skew: {skewness:.2f}',
                    xy=(0.7, 0.9),
                    xycoords='axes fraction',
                    fontsize=10,
                    color='red')

        ax.set_title(column)
        ax.set_xlabel('Değerler')
        ax.set_ylabel('Frekans')
        ax.grid(True)

    plt.tight_layout()
    plt.show()

# Fonksiyonu çağırarak grafikleri çizdirin
plot_histogram_with_skew(df_temp)
```
